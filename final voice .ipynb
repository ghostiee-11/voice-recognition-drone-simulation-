```python
import speech_recognition as sr
import pyttsx3


engine = pyttsx3.init()

def speak(text):
    """Converts text to speech."""
    engine.say(text)
    engine.runAndWait()

def drone_move(command):
    """Executes drone commands based on voice input."""
    if "forward" in command:
        print("Moving drone forward.")
        speak("Moving drone forward.")
        
    elif "backward" in command:
        print("Moving drone backward.")
        speak("Moving drone backward.")
        
    elif "left" in command:
        print("Turning drone left.")
        speak("Turning drone left.")
       

    elif "right" in command:
        print("Turning drone right.")
        speak("Turning drone right.")
        

    elif "upward" in command:
        print("Drone moving up.")
        speak("Drone moving up.")
       

    elif "downward" in command:
        print("Drone moving down.")
        speak("Drone moving down.")
        

    elif "stop" in command:
        print("Stopping drone.")
        speak("Stopping drone.")
       

    elif "flip" in command:
        print("Flipping drone.")
        speak("Flipping drone.")
       

    elif "throttle" in command:
        print("Throttling drone.")
        speak("Throttling drone.")
       

    else:
        print("Command not recognized.")
        speak("Command not recognized.")

def recognize_command():
    """Recognizes voice commands and passes them to the drone_move function."""
    recognizer = sr.Recognizer()
    microphone = sr.Microphone()

    with microphone as source:
        print("Adjusting for ambient noise, please wait...")
        recognizer.adjust_for_ambient_noise(source)
        print("Listening for a command...")
        speak("Listening for a command.")
        
        try:
            audio = recognizer.listen(source, timeout=5)
            command = recognizer.recognize_google(audio).lower()
            print(f"Recognized command: {command}")
            drone_move(command)

        except sr.WaitTimeoutError:
            print("Listening timed out. No command detected.")
            speak("Listening timed out. No command detected.")
        except sr.UnknownValueError:
            print("Sorry, I didn't catch that.")
            speak("Sorry, I didn't catch that.")
        except sr.RequestError as e:
            print(f"Could not request results from the recognition service; {e}")
            speak("Could not request results from the recognition service.")

# Main loop 
if __name__ == "__main__":
    while True:
        recognize_command()
```
